Value string to number

```python
import numpy as np
def f(x):
    return float(x.replace('$','').replace('.',''))

df['value'] = np.vectorize(f)(df['value'])
```

Split Goalkeeper

```python
df_gk = df[['gk_positioning', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_reflexes', 'value']]
df_gk
```

```python
df_gk = df_gk[(df_gk['gk_positioning'] >= 40) | (df_gk['gk_diving'] >= 40) | (df_gk['gk_handling'] >= 40) | (df_gk['gk_kicking'] >= 40) | (df_gk['gk_reflexes'] >= 40)]
df_gk
```

# Regression

```python
from pycaret.regression import *
```

```python
setup(df_gk, target='value')
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
save_model(best_model, 'model_reg_goalkeeper')
```
