# Clustering

```python
from pycaret.clustering import *
```

```python
df.columns
```

```python
physical = df[['acceleration', 'agility', 'balance',
            'jumping', 'reactions', 'sprint_speed',
            'stamina', 'strength']]

mental = df[['aggression', 'composure', 'interceptions', 'att_position', 'vision']]

skill = df[['ball_control', 'crossing', 'curve',
             'dribbling', 'finishing', 'fk_acc',
             'heading', 'long_pass', 'long_shots',
             'penalties', 'shot_power', 'short_pass',
             'slide_tackle', 'stand_tackle', 'volleys']]

gk = df[['gk_positioning', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_reflexes']]
```

### Physical

```python
setup(physical)
```

```python
physical_model = create_model('kmeans', num_clusters=3)
```

```python
evaluate_model(physical_model)
```

```python
save_model(physical_model, 'model_physical')
```
